```python
import pandas as pd

# Load your data
df = pd.read_csv('combined.csv')

# Clean and extract departure hour from DepTime (e.g., 1830 â†’ 18)
df = df[df['DepTime'].notna()]
df['DepTime'] = df['DepTime'].astype(int)
df['DepHour'] = df['DepTime'].astype(str).str.zfill(4).str[:2].astype(int)

# Define delayed flights (more than 15 minutes delay)
df['IsDelayed'] = df['ArrDelay'] > 15

# Calculate overall delay rate
overall_delay_rate = df['IsDelayed'].mean()

# Filter for flights departing between 6 PM (18) and 10 PM (21)
evening_flights = df[(df['DepHour'] >= 18) & (df['DepHour'] <= 21)]
evening_delay_rate = evening_flights['IsDelayed'].mean()

# Calculate percentage increase
increase = ((evening_delay_rate - overall_delay_rate) / overall_delay_rate) * 100

# Print result
print(f"Evening delay rate: {evening_delay_rate:.2%}")
print(f"Overall delay rate: {overall_delay_rate:.2%}")
print(f"Increase in delay: {increase:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("combined.csv")  # Use the correct path if different

# Set the style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 1. Heatmap: Average Arrival Delay by Day of Week and Month
heatmap_data = df.pivot_table(values='ArrDelay', index='DayOfWeek', columns='Month', aggfunc='mean')
plt.subplot(3, 1, 1)
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm")
plt.title('Average Arrival Delay by Day of Week and Month')

# 2. Bar Chart: Average Arrival Delay by Airline
plt.subplot(3, 1, 2)
carrier_delay = df.groupby('UniqueCarrier')['ArrDelay'].mean().sort_values()
sns.barplot(x=carrier_delay.index, y=carrier_delay.values, palette="viridis")
plt.ylabel('Average Arrival Delay')
plt.xlabel('Airline')
plt.title('Average Arrival Delay by Airline')

# 3. Box Plot: Departure Delay Distribution by Month
plt.subplot(3, 1, 3)
sns.boxplot(x='Month', y='DepDelay', data=df)
plt.title('Departure Delay Distribution by Month')
plt.ylabel('Departure Delay (minutes)')
plt.xlabel('Month')

# Adjust layout
plt.tight_layout()
plt.show()
```
